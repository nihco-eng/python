```python
import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic')
plt.rc('axes', unicode_minus=False)

import pandas as pd
df = pd.read_csv('data/score.csv')
df
```

```python
x= df.index * 4
plt.figure(figsize=(10, 5))
plt.bar(x, df['국어'], width=1, label='국어')
plt.bar(x+1, df['영어'], width=1, label='영어')
plt.bar(x+2, df['수학'], width=1, label='수학')
plt.legend()
xticks = [xtick+1 for xtick in x]
plt.xticks(xticks, df['이름'])
plt.show()
```

```python
x =df.index * 4
plt.bar(x + 0, df['국어'], label='국어', width=1)
plt.bar(x + 1, df['영어'], label='영어', width=1)
plt.bar(x + 2, df['수학'], label='수학', width=1)
xticks = [idx+1 for idx in x]
plt.xticks(xticks, df['이름'])
for idx in df.index:
    x= idx *4
    plt.text(x, df.loc[idx, '국어']/2, df.loc[idx, '국어'], ha='center', size=8, color='white')
    plt.text(x+1, df.loc[idx, '영어']+1, df.loc[idx, '영어'], ha='center', size=8)
    plt.text(x+2, df.loc[idx, '수학']+1, df.loc[idx, '수학'], ha='center', size=8)
plt.legend()
plt.show()
```

```python
x = df.index * 3
plt.bar(x, df['사회'], label='사회', width=1)
plt.bar(x + 1, df['과학'], label='과학', width=1)
xticks = [idx+0.5 for idx in x]
plt.xticks(xticks, df['이름']+'\n'+ df['지원번호'])
for idx in df.index:
    x= idx *3
    plt.text(x, df.loc[idx, '사회']/2, df.loc[idx, '사회'], ha='center', size=8, color= 'white')
    plt.text(x+1, df.loc[idx, '과학']+1, df.loc[idx, '과학'], ha='center', size=8)
plt.legend()
plt.show()
```
