```python
import pandas as pd
df = pd.read_csv('data/score.csv', index_col='지원번호')
df

import matplotlib.pyplot as plt

#한글 설정
plt.rc('font', family = 'Malgun Gothic')
plt.rc('font', size = 10)
plt.rc('axes', unicode_minus = False)
```

```python
import numpy as np
def graph(name):
    filt = df['이름'] == name
    df2 = df[filt]
    if len(df2) == 0 :
        return np.nan
    else :
        df3 = df2.loc[:,['국어','과학','사회']]
        df4 = df3.iloc[0]
        plt.bar(df4.index, df4.values)
        return plt
image = graph('변덕규')
if pd.isna(image):
    print('없음')
else:
    plt.show() 
```

```python
name = df['이름']
height = df['키']
kor = df['국어']
eng = df['영어']
mat = df['수학']
si = df['과학']
so = df['사회']
mean = (kor+eng+mat+si+so)/5
```

```python
plt.barh(name, height)
plt.ylabel('학생이름')
plt.xlabel('키')
plt.xlim(160, 220)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
bar = plt.bar(name, height, color='none', ec = 'g')
plt.ylim(160, 210)
bar[0].set_hatch('/')
for idx, h in enumerate(height):
    plt.text(idx, h+2, h, ha='center')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.title('학생별 평균')
bar = plt.bar(name, mean, ec = 'purple')
plt.ylim(0, 100)
plt.yticks([y for y in range(0, 101, 5)])
plt.ylabel('학생이름')
plt.xlabel('평균점수')
for idx, h in enumerate(mean):
    plt.text(idx, h+2, h, ha='center')
plt.show()
```
