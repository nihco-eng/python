```python
import pandas as pd
df = pd.read_csv('data/score.csv', index_col='지원번호')
df

import matplotlib.pyplot as plt

#한글 설정
plt.rc('font', family = 'Malgun Gothic')
plt.rc('font', size = 10)
plt.rc('axes', unicode_minus = False)
```

```python
kor = list(df['국어'])
eng = list(df['영어'])
mat = list(df['수학'])
name = df['이름']
```

```python
plt.figure(figsize=(10,5))
plt.plot(name, kor, label='국어', c='red', marker='o', ls='--')
plt.plot(name, eng, label='영어', c='blue', marker='s', ls='-.')
plt.plot(name, mat, label='수학', c='green', marker='X')
plt.legend(loc=(1.02, 0.72))
for idx, k in enumerate(kor):
    plt.text(name[f'{idx+1}번'], kor[idx]+2, k, ha='center', size=10) 
plt.show()
```
