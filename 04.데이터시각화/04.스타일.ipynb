```python
import pandas as pd
df = pd.read_csv('data/score.csv', index_col='지원번호')
df

import matplotlib.pyplot as plt

#한글 설정
plt.rc('font', family = 'Malgun Gothic')
plt.rc('font', size = 10)
plt.rc('axes', unicode_minus = False)
```

```python
name = list(df['이름'])
height = list(df['키'])
name, height
```

```python
plt.title('학생별 신장\n', size=20)
plt.xlabel('이름', size =12)
plt.ylabel('키(cm)', size= 12)
plt.yticks([y for y in range(160, 220, 5)])
plt.plot(name, height, linewidth='2', linestyle= '--', color='red', marker= 'o', markersize =10, markerfacecolor = 'yellow', markeredgecolor='red', markeredgewidth=2)
plt.show()
```

```python
plt.title('학생별 신장\n', size=20)
plt.xlabel('이름', size =12)
plt.ylabel('키(cm)', size= 12)
plt.yticks([y for y in range(160, 220, 5)])
plt.plot(name, height, lw=5, c='red', ls='--', marker = 'o', ms=20, mfc='#7B68EE', mec='#191970', mew=5)
plt.show()
```

```python
plt.figure(figsize=(15, 3))
plt.ylim(160, 220)
plt.plot(name, height, marker='o')
for idx, txt in enumerate(height) :
    plt.text(name[idx], height[idx]-4, f'{txt}cm', size=8, ha='center')
plt.show()
```
