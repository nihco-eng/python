```python
import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic')
plt.rc('axes', unicode_minus=False)
```

```python
import pandas as pd
df = pd.read_csv('data/score.csv')
df
```

```python
group = df.groupby('SW특기').size()
plt.pie(group.values, explode=[0.05] * len(group), wedgeprops={'width':0.8, 'alpha':0.3, 'edgecolor':'lightpink', 'linewidth':5}, 
        labels=group.index, autopct="%.1f%%", textprops={'color':'pink', 'size':10})
plt.legend(loc=(1, 0.5))
plt.show()
```

```python
import pandas as pd
enroll = pd.read_csv('data/수강.csv')
enroll
```

```python
group = enroll.groupby('강좌번호').size().to_frame('인원수')
cou = pd.read_csv('data/강좌.csv')
merge = cou.merge(group, on='강좌번호')
merge
```

```python
colors = ['#FFADAD','#FFD6A5','#FDFFB6','#CAFFBF','#9BF6FF','#A0C4FF']
labels = merge['강좌이름']
values = merge['인원수']
plt.pie(values, labels=labels, colors = colors, wedgeprops ={'width':0.8}, autopct='%.1f%%', explode=[0.05]*len(values))
plt.show()
```

```python
stu = pd.read_csv('data/학생.csv')
group = stu.groupby('학생학과').size()
```

```python
labels = group.index
values = group.values
plt.pie(values, labels=labels, autopct='%.0f%%')
plt.show()
```

```python
def grade(score):
    if score >= 90:
        return 'A'
    elif score >=80:
        return 'B'
    elif score >=70:
        return 'C'
    elif score >=60:
        return 'D'
    else :
        return 'F'
```

```python
enroll = pd.read_csv('data/수강.csv')
enroll['학점'] = enroll['점수'].apply(grade)
group = enroll.groupby('학점').size()
group
```

```python
plt.pie(group.values, labels=group.index+'학점', autopct='%.0f%%', explode=[0.05]*len(group), wedgeprops={'width':0.8}, textprops={'color':'pink', 'size':10})
plt.show()
```
