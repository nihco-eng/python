```python
import numpy as np
from PIL import Image
```

```python
heart_mask = np.array(Image.open('heart_mask.png'))
heart_mask
```

```python
plt.figure(figsize=(5, 5))
wc= WordCloud(width=800, height=800, background_color='white', min_font_size=20, mask=heart_mask).generate(text)
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.savefig('골든하트.png')
plt.show()
```

```python
from konlpy.tag import Okt
from collections import Counter
```

```python
with open('애국가.txt', 'r', encoding='utf-8') as file:
    text = file.read()
print(text)
```

```python
with open('애국가.txt', 'r', encoding='utf-8') as file:
    text = file.read()
print(text)
```

```python
okt = Okt()
nouns =okt.nouns(text)
nouns = [noun for noun in nouns if len(noun) > 1]
nouns
```

```python
counter = Counter(nouns)
counter
```

```python
wc= WordCloud(width=800, height=800, background_color='white', scale=2.0, max_font_size=250, font_path='MALGUNBD.TTF', mask=heart_mask).generate_from_frequencies(counter)
```

```python
plt.imshow(wc)
plt.axis('off')
plt.savefig('애국가.png')
plt.show()
```
