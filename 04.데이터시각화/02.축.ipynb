```python
import matplotlib.pyplot as plt
#한글 설정
plt.rc('font', family = 'Malgun Gothic')
plt.rc('font', size = 10)
plt.rc('axes', unicode_minus = False)
```

```python
x = [1, 2, 3, 4, 5]
y = [3, 6, 9, 12 ,15]
```

```python
plt.plot(x, y)
plt.xticks([x for x in range(min(x), max(x)+1, 1)], size =15, color= 'blue')
plt.yticks([y for y in range(min(y), max(y)+1, 1)], size =15, color= 'green')
plt.show()
```
