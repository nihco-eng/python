```python
import pandas as pd
df = pd.read_csv('data/score.csv', index_col='지원번호')
df

import matplotlib.pyplot as plt

#한글 설정
plt.rc('font', family = 'Malgun Gothic')
plt.rc('font', size = 10)
plt.rc('axes', unicode_minus = False)
```

```python
group = df.groupby('학교')['키'].mean()
group
```

```python
school = group.index
height = group.values
school, height
```

```python
plt.bar(school, height)
for idx, h in enumerate(height):
    plt.text(idx, h+2, f'{h:.2f}cm', ha='center')
plt.xticks(school, [s+'등학교' for s in school], rotation = 45)
plt.show()
```

```python
#학생별 키 막대그래프

import pandas as pd
df = pd.read_csv('c:/python/04.데이터시각화/data/score.csv', index_col='지원번호')
df

import matplotlib.pyplot as plt

#한글 설정
plt.rc('font', family = 'Malgun Gothic')
plt.rc('font', size = 10)
plt.rc('axes', unicode_minus = False)
```

```python
df
```

```python
name = df['이름']
height = df['키']
plt.figure(figsize=(10, 3))
plt.ylim(150, 210)
plt.bar(name, height, color='purple', hatch='/', ec='green')
plt.xticks(name, rotation=45, size=8, color='pink')
for idx, h in enumerate(height):
    plt.text(idx, h+1, h, ha='center', color='skyblue', size=8)
```

```python
df['평균'] = df.apply(lambda m:m['국어':'사회'].mean(), axis=1)
name = df['이름']
avg = df['평균']
plt.figure(figsize=(10, 5))
plt.ylim(0, 100)
plt.bar(name, avg, color='purple', hatch='/', ec='green')
plt.xticks(name, rotation=45, size=8, color='pink')
for idx, h in enumerate(avg):
    plt.text(idx, h+1, f'{h:.2f}', ha='center', color='skyblue', size=8)
```

```python
#학교별 평균키
group = df.groupby('학교')['키'].mean()
labels = group.index
values = group.values
plt.figure(figsize=(10, 5))
plt.ylim(160, 200)
plt.bar(labels, values, color='purple', hatch='/', ec='green', width=0.3)
plt.xticks(labels, rotation=45, size=8, color='pink')
for idx, value in enumerate(values):
    plt.text(idx, value+1, f'{value:.2f}cm', ha='center', color='skyblue', size=8)
```

```python
#학교별 평균점수
df['평균'] = df.apply(lambda m:m['국어':'사회'].mean(), axis=1)
group = df.groupby('학교')['평균'].mean()
labels = group.index
values = group.values
plt.figure(figsize=(10, 5))
plt.ylim(0, 100)
plt.bar(labels, values, color='purple', hatch='/', ec='green', width=0.3)
plt.xticks(labels, rotation=45, size=8, color='pink')
for idx, value in enumerate(values):
    plt.text(idx, value+1, f'{value:.2f}점', ha='center', color='skyblue', size=8)
```

```python
#SW특기별 인원수
df['SW특기'] = df['SW특기'].str.capitalize()
group = df.groupby('SW특기').size()
labels = group.index
values = group.values
plt.figure(figsize=(10, 5))
plt.ylim(0, 6)
plt.bar(labels, values, color='purple', hatch='/', ec='green', width=0.3)
plt.xticks(labels, rotation=45, size=10, color='pink')
for idx, value in enumerate(values):
    plt.text(idx, value+0.09, f'{value}명', ha='center', color='skyblue', size=8)
```

```python
#학생별(이름) 평균점수 
enroll = pd.read_csv('data/수강.csv')
group = enroll.groupby('학생번호')['점수'].mean().to_frame('평균점수')
stu = pd.read_csv('data/학생.csv')
merge = stu.merge(group, how='left', on='학생번호')
merge.fillna(0, inplace=True)
merge
```

```python
#학생별 평균점수 막대그래프
stu = merge['학생이름']
grade = merge['평균점수']
nos = merge['학생번호']
plt.bar(stu, grade)
plt.ylim(0, 100)
yticks = [y for y in range(0, 101, 20)]
plt.yticks(yticks, [f'{y}점' for y in yticks])
xticks = [f'{stu[idx]}({nos[idx]})' for idx in range(len(nos))]
plt.xticks(stu, xticks, rotation=45)
for idx, grade in enumerate(merge['평균점수']):
    plt.text(idx, grade+0.9, f'{grade:.2f}', ha='center')
plt.show()
```

```python
#강좌별(이름) 평균 막대그래프
cou = pd.read_csv('data/강좌.csv')
enroll = pd.read_csv('data/수강.csv')
group = enroll.groupby('강좌번호')['점수'].mean().to_frame('평균점수')
merge = cou.merge(group, how='left', on='강좌번호')
nos = merge['강좌번호']
names = merge['강좌이름']
grades = merge['평균점수']
merge.fillna(0, inplace=True)
plt.bar(names, grades)
plt.ylim(0, 100)
yticks = [y for y in range(0, 101, 20)]
plt.yticks(yticks, [f'{y}점' for y in yticks])
xticks = [f'{names[idx]}({nos[idx]})' for idx in range(len(nos))]
plt.xticks(names, xticks, rotation=45)
for idx, grade in enumerate(merge['평균점수']):
    plt.text(idx, grade+0.9, f'{grade:.2f}점', ha='center')
plt.show()
```
