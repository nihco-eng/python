```python
#성별 생존율
total = train.groupby('Sex').size().to_frame('전체인원')
total
```

```python
filt = train['Survived'] == 1
survived = train[filt]
survived = survived.groupby('Sex').size().to_frame('생존인원')
survived
```

```python
gender = total.merge(survived, on='Sex')
gender['생존율(%)'] = gender['생존인원']/gender['전체인원'] * 100
gender
```

```python
#티켓 클래스별 생존율
total = train.groupby('Pclass').size().to_frame('전체인원')
total
```

```python
filt = train['Survived'] == 1
survived = train[filt]
pclass = survived.groupby('Pclass').size().to_frame('생존인원')
pclass
```

```python
pclass = total.join(pclass)
pclass['생존율(%)'] = pclass['생존인원']/pclass['전체인원']
pclass
```

```python
train['Age'].describe()
```

```python
min = int(train['Age'].min())
max = int(train['Age'].max())
min, max
```

```python
ages = []
step = 20
for age in range(min, max+1, step):
    filt = (train['Age'] >= age) & (train['Age'] < age+step)
    total = train[filt]
    count = train[filt]['Age'].count() #연령대별 전체인원
    filt = (train['Age'] >= age) & (train['Age'] < age+step) & (train['Survived']==1)
    count_survived = train[filt]['Age'].count() #연령대별 생존자수
    ages.append({
                '나이':f'{age}이상 ~ {age+step}미만', 
                '전체인원': count,
                '생존인원':count_survived})
ages

df = pd.DataFrame(ages)
df.set_index('나이', inplace=True)
df['생존율'] = df['생존인원']/df['전체인원']
df
```

```python
#탑승항구(Embarked)별 전체인원, 생존인원, 생존율
ages = []
step = 20
for age in range(min, max+1, step):
    filt = (train['Age'] >= age) & (train['Age'] < age+step)
    total = train[filt]
    count = train[filt]['Age'].count() #연령대별 전체인원
    filt = (train['Age'] >= age) & (train['Age'] < age+step) & (train['Survived']==1)
    count_survived = train[filt]['Age'].count() #연령대별 생존자수
    ages.append({
                '나이':f'{age}이상 ~ {age+step}미만', 
                '전체인원': count,
                '생존인원':count_survived})
ages

df = pd.DataFrame(ages)
df.set_index('나이', inplace=True)
df['생존율'] = df['생존인원']/df['전체인원']
df
```

```python
#티켓 클래스별 생존율
total = train.groupby('Embarked').size().to_frame('전체인원')
filt = train['Survived'] == 0
survived = train[filt]
Embarked = survived.groupby('Embarked').size().to_frame('사망인원')
Embarked = total.join(Embarked)
Embarked['사망률(%)'] = Embarked['사망인원']/Embarked['전체인원']
Embarked
```
