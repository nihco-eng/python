```python
def add_cm(height) :
    if isinstance(height, int) :
        return f'{height}cm'
    else : 
        return int(height.replace('cm', ''))
```

```python
df2 = df.copy()
df2['키'] = df['키'].apply(add_cm)
df2
```

```python
df2['키'] = df2['키'].apply(add_cm)
df2
```

```python
df3 = df.copy()
df3
```

```python
df3['키'] = df3['키'].apply(lambda x: f'{x}cm' if isinstance(x, int) else int(x.replace('cm', '')))
df3
```

```python
df4 = df.copy()
df4['SW특기'] = df4['SW특기'].str.lower()
df4
```

```python
def lower(sw):
    if pd.notnull(sw) :
        return sw.lower()
    else : 
        return '없음'
```

```python
df4['SW특기'] = df4['SW특기'].apply(lower)
df4
```

```python
df4['SW특기'] = df4['SW특기'].apply(
    lambda x : x.lower() if pd.notnull(x) else '없음'
)
df4
```

```python
df_score = df.loc[:,'국어':'사회']
df_score['총점'] = df_score.apply(
    lambda row :sum(row['국어':'사회']), axis=1)
df_score['평균'] = df_score.apply(
    lambda row :row['총점']/5, axis=1)
df_score.loc['평균'] = df_score.apply(np.mean, axis=0)
df_score
```
