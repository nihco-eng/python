```python
import pandas as pd
```

```python
df_info = pd.read_csv('data/학생정보.csv')
df_info
```

```python
df_score = pd.read_csv('data/학생성적.csv')
df_score
```

```python
idx = len(df_score)
no = f'{idx+1}번'
no
```

```python
df_score.loc[idx]=[no, 0, 0, 0, 0, 0]
df_score
score = df_score
```

```python
filt = score['지원번호'] == '8번'
idx = score[filt].index
if len(idx) == 0 : 
    print('없습니다.')
else:
    score.drop(index=idx, inplace=True)
    print('삭제완료')
score
```

```python
df = df_info.merge(df_score, how = 'left', left_on ='지원번호', right_on = '지원번호1')
df.drop(columns='지원번호1', inplace = True)
df.fillna({'SW특기':''}, inplace = True)
df.fillna(0, inplace= True)
df
```
